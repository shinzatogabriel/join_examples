```
peopledf = spark.createDataFrame([
  ("John","01"),
  ("Fred","01"),
  ("Harry","02"),
  ("Sally","04"),
  ("Anna","06"),
  ("Rachel","01"),
  ("Bob","03")
  ],["Person","Country"])
        
countrydf = spark.createDataFrame([
  ("01","UK"),
  ("02","France"),
  ("03","Germany"),
  ("04","Spain"),
  ("05","Belgium")
  ],["CountryKey","Country"])

peopledf.display()
countrydf.display()
```

```
# INNER
innerdf = peopledf.join(countrydf, peopledf.Country == countrydf.CountryKey)
display(innerdf)
```

```
leftdf = peopledf.join(countrydf, peopledf.Country == countrydf.CountryKey, 'left')
display(leftdf)
```

```
rightdf = peopledf.join(countrydf, peopledf.Country == countrydf.CountryKey, 'right')
display(rightdf)
```

```
leftsemidf = peopledf.join(countrydf, peopledf.Country == countrydf.CountryKey, 'leftsemi')
display(leftsemidf)
```

```
leftantidf = peopledf.join(countrydf, peopledf.Country == countrydf.CountryKey, 'leftanti')
display(leftantidf)
```

```
crossdf = peopledf.crossJoin(countrydf)
display(crossdf)
```
